```python
mtcars = sns.load_dataset('mpg')

# Carregando o banco de dados mtcars
# O conjunto de dados mpg é um equivalente aproximado ao mtcars de R
# Conjunto de dados mpg (miles per gallon) é um exemplo frequentemente utilizado para análise estatística e visualização em Python, especialmente na biblioteca Seaborn.
# Ele é semelhante ao famoso conjunto de dados mtcars do R, sendo frequentemente usado para demonstrações de manipulação e visualização de dados.

print(mtcars)

# Exibe os dados
```

```python
cylinder_counts = mtcars['cylinders'].value_counts()

# Contando o número de carros para cada número de cilindros

# mtcars é um DATAFRAME, ou seja, uma estrutura de dados do pandas (biblioteca Python usada para manipulação de dados tabulares),
# que pode ser imaginado como uma tabela com linhas e colunas, onde cada coluna pode ter um tipo diferente de dado (números, texto, datas, etc.)

# Vamos acessar a coluna chamada "cylinders" dentro do DataFrame mtcars, assa coluna contém o número de cilindros dos carros
# ou seja, a quantidade de cilindros no motor de cada carro listado.

# O value_counts() é uma função do pandas que conta quantas vezes cada valor único aparece em uma coluna específica.
# Neste caso, ele vai contar quantos carros têm 4 cilindros, 6 cilindros, 8 cilindros, etc.

print(cylinder_counts)

# Exibe os dados
# Note que o resultado de uma contagem de frequência dos carros com diferentes números de cilindros, similar ao uso do método value_counts(),
# mas com um nome específico para a série.

# A Parte Name: count, dtype: int64:
# Name: count: Isso indica que a série de dados tem o nome count. Esse nome pode ter sido dado ao aplicar uma função de agregação ou pode ser o nome de uma
# coluna do DataFrame que foi renomeada.

# dtype: int64: Isso informa que o tipo de dado (data type) dos valores dessa série é inteiro de 64 bits. Ou seja,
# os números de carros contados são armazenados como inteiros (valores numéricos sem decimais).
```

```python
#ESTATÍSTICA DESCRITIVA

# Estatística descritiva é um conjunto de técnicas utilizadas para organizar, resumir e descrever
# as características de um conjunto de dados - essencial para entender as informações antes de fazer
# análises mais profundas. No Python, podemos realizar estatísticas descritivas usando diversas bibliotecas
# como o NumPy, Pandas e SciPy.
```

```python
# Calcular a média amostral da coluna "cylinder"

# A média amostral é a média dos valores de uma amostra de dados, e é uma medida de tendência central
# da amostra.
mean_cyl = np.mean(mtcars['cylinders'])
print(f'Média amostral de cylinders: {mean_cyl}')
```

```python
# Calcular a variância amostral da coluna "cylinder"

# Variância amostral é uma medida que quantifica o grau de dispersão ou variação dos dados
# em torno da média amostral. Ela é o quadrado do desvio padrão e indica o quanto os valores
# individuais de uma amostra se afastam da média.

variance_cyl = np.var(mtcars['cylinders'], ddof=1)
print(f'Variância amostral de cylinders: {variance_cyl}')
```

```python
# Calcular o resumo de cinco pontos para a coluna 'cylinder'
summary = mtcars['cylinders'].describe()
print("\nResumo de cinco pontos + Média:")
print(summary)
```

```python
# Calculando as frequências absolutas para a coluna 'cyl'
frequencias_absolutas = mtcars['cylinders'].value_counts()

# Criando o gráfico de barras com cores personalizadas e bordas pretas
ax = frequencias_absolutas.plot(kind='bar',
                                color=plt.cm.rainbow(np.linspace(0, 1, len(frequencias_absolutas))),  # Cores do arco-íris
                                edgecolor='black')  # Bordas pretas

# Ajustando o limite do eixo Y
plt.ylim(0, max(frequencias_absolutas) + 2)

# Adicionando título e rótulos aos eixos
plt.title('Número de cilindros', fontsize=16, fontweight='bold')
plt.xlabel('Cilindros', fontsize=14, fontweight='bold')
plt.ylabel('Frequência', fontsize=14, fontweight='bold')

# Ajustando o tamanho das fontes dos rótulos
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Exibindo o gráfico
plt.show()
```

```python
# Criando o gráfico de pizza (setores)
plt.pie(frequencias_absolutas,
        labels=['4 c', '8 c', '6 c', '3 c', '5 c' ],  # Rótulos dos setores
        colors=['yellow', 'red', 'blue', 'orange', 'brown'],  # Cores dos setores
        autopct='%1.1f%%',  # Exibe a porcentagem de cada setor
        startangle=90,  # Inicia o gráfico de pizza de forma rotacionada
        wedgeprops={'edgecolor': 'black'})  # Borda preta nas fatias

# Adicionando título
plt.title('Distribuição de Cilindros')

# Exibindo o gráfico
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D #pizzaem 3d
```

```python
# Dados para o gráfico de pizza
sizes = [51.3, 25.9, 21.1, 1.0, 1.7 ]  # Exemplo de percentuais
labels = ['4 c', '8 c', '6 c', '3 c', '5 c']
colors = ['yellow', 'red', 'blue', 'orange', 'brown']

# Criando o gráfico de pizza em 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Angulos para os setores
angles = np.linspace(0, 2 * np.pi, len(sizes), endpoint=False)

# Convertendo para coordenadas 3D
x = np.cos(angles) * sizes
y = np.sin(angles) * sizes
z = np.zeros_like(x)

# Criando as fatias do gráfico de pizza
ax.bar3d(x, y, z, np.ones_like(x), np.ones_like(y), sizes, color=colors, zsort='average')

# Adicionando rótulos aos setores
for i in range(len(sizes)):
    ax.text(x[i], y[i], sizes[i] / 2, labels[i], horizontalalignment='center')

# Definindo título e visualizando o gráfico
ax.set_title('Distribuição de Cilindros (Pizza 3D)')
plt.show()
```

```python
#testando aderência à normalidade com teste de shapiro
```

```python
from scipy.stats import shapiro
# Realizando o teste de Shapiro-Wilk para verificar a normalidade
stat, p_value = shapiro(mtcars['horsepower'].dropna())

# Exibindo os resultados
print("Estatística do teste:", stat)
print("Valor p:", p_value)

# Interpretação do valor p
if p_value > 0.05:
    print("A distribuição parece ser normal (não rejeitamos H0)")
else:
    print("A distribuição não é normal (rejeitamos H0)")
```
